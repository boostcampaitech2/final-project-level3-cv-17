```python
import os
from os.path import splitext, join
import pandas as pd
import numpy as np
from PIL import Image
from tqdm import tqdm

img_root = r'E:\food_raw\Training\stew\train'
img_dirs = os.listdir(img_root)

# -- label data 2 dataframe & xywh 2 xyxy
df = pd.DataFrame([], columns=['mean1', 'mean2', 'mean3', 'std1', 'std2', 'std3'])

for img_dir in tqdm(img_dirs):
    for img_file in os.listdir(join(img_root, img_dir)):
        cur_img = np.array(Image.open(join(img_root, img_dir, img_file)))

        mean1, mean2, mean3 = cur_img.mean(axis=(0,1))
        std1, std2, std3 = cur_img.std(axis=(0,1))

        df = df.append({'mean1':mean1, 'mean2':mean2, 'mean3':mean3, 'std1':std1, 'std2':std2, 'std3':std3}, ignore_index=True)
```

```python
print([round(val, 2) for val in list(df.describe().loc['mean'] / 255.)])
```
